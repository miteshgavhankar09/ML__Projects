OHE Encoder

```python
# Encode the features

X = horse_df.drop("outcome" ,axis =1)
y = horse_df["outcome"]
```

```python
X_encoded = pd.get_dummies(horse_df)
X_encoded
```

```python
X_encoded.shape
```

```python
X.shape
```

```python
from sklearn.preprocessing import LabelEncoder

horse_outcome_encoder =LabelEncoder()

y = horse_outcome_encoder.fit_transform(y)

print(y)
```

We can infer from the principle of a label encoder that

1. 0 Represents "died"
2. 1 Represents "euthanized"
3. 2 Represents "lived"

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_encoded , y ,test_size = 0.2, random_state =21)
```

```python
print("Shape of X_train is : ", X_train.shape)
print("Shape of X_test is : ", X_test.shape)
print("Shape of y_train is : ", y_train.shape)
print("Shape of y_test is : ", y_test.shape)
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
imp = SimpleImputer(missing_values = np.nan, strategy = "most_frequent")
X_train = imp.fit_transform(X_train)
X_test = imp.fit_transform(X_test)
```

```python
pd.DataFrame(X_train).isnull().sum()
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc_model = RandomForestClassifier(n_estimators = 150) # n_estimators refers to the number of decision tree used. 
```

```python
rfc_model.fit(X_train,y_train)
```

```python
y_pred_test =rfc_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy score of the RFC model on test dataset is: ")
print(accuracy_score(y_test,y_pred_test))
```

# Train Data Accuracy

```python
y_pred_train =rfc_model.predict(X_train)

print("Accuracy score of the RFC model on test dataset is: ")
print(accuracy_score(y_train,y_pred_train))
```

```python
# Pruning the RFC
rfc_pruned_model = RandomForestClassifier(n_estimators=150,max_depth = 5) # Experimenting max depth parameter.
```

```python
y_pred_test = rfc_pruned_model.fit(X_train,y_train)
```

```python
y_pred_test =rfc_pruned_model.predict(X_test)
```

```python
print("Accuracy score of the RFC model on test dataset is: ")
print(accuracy_score(y_test,y_pred_test))
```

```python
y_pred_train =rfc_pruned_model.predict(X_train)
```

```python
print("Accuracy score of the RFC model on train dataset is: ")
print(accuracy_score(y_train,y_pred_train))
```

# Adaboost Classifier

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ada_model = AdaBoostClassifier(n_estimators = 150) # base_estimator = none means by default it will take decision tree.

ada_model.fit(X_train,y_train)

y_pred_ada_test =ada_model.predict(X_test)

y_pred_ada_train =ada_model.predict(X_train)
```

```python
print("Accuracy score of the AdaBoost model on test dataset is: ")
print(accuracy_score(y_test,y_pred_ada_test))
```

```python
print("Accuracy score of the AdaBoost model on train dataset is: ")
print(accuracy_score(y_train,y_pred_ada_train))
```
